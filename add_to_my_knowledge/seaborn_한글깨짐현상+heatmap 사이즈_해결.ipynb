# matplotlib과 seaborn으로 시각화를 할 때 한글이 깨지는 현상

- 어제 '상관분석'을 하게 되어서, 아래의 블로그를 들어가보게 되었다.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test = pd.DataFrame(np.random.randint(0,100,size = (15,15)), columns = list('가나다라마바사아자차카타파하감'), index = list('가갸거겨고교구규그기나냐너녀노'))
```

```python
test
```

```python
df_heatmap = sns.heatmap(test, cbar = True, annot = True, annot_kws = {'size':20}, square = True, cmap = 'Reds')
```

- 이렇게 만들어버리면, 네모 칸 안에 있는 글씨는 글씨대로 난리가 나고 바깥에 있는 한글은 다 깨지게 된다.
